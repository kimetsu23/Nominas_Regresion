```python
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
DF_Nomina_Agricultura_Bruta = pd.read_csv("Agricultura_Nomina.csv")
```

```python
# Copia TODO esto en una celda y ejecuta:
DF_Nomina_Agricultura_NETA = DF_Nomina_Agricultura_Bruta.copy()


# 1. Renombrar columnas necesarias
DF_Nomina_Agricultura_NETA = DF_Nomina_Agricultura_NETA.rename(columns={
    'Unnamed: 1': 'Nombres',
    'Unnamed: 2': 'Apellidos',
    'Unnamed: 3': 'Cargo_Funcion',
    'Unnamed: 4': 'Locacion_Departamento',
    'Unnamed: 7': 'Salario_Bruto',
    'Unnamed: 8': 'Descuento_AFP',
    'Unnamed: 9': 'Descuento_ISR',
    'Unnamed: 10': 'Descuento_SFS',
    'Unnamed: 16': 'Total_Descuentos',
    'Unnamed: 17': 'Salario_Neto'
})


# 2. Crear UNA SOLA columna con nombre y apellido combinados
DF_Nomina_Agricultura_NETA['Nombre_Completo'] = DF_Nomina_Agricultura_NETA['Nombres'] + ' ' + DF_Nomina_Agricultura_NETA['Apellidos']


# 3. Crear columna de otros descuentos (ya como números)
# NOTA: Añadimos 'r' antes de las comillas para raw strings
DF_Nomina_Agricultura_NETA['Otros_Descuentos'] = (
    pd.to_numeric(DF_Nomina_Agricultura_NETA['Unnamed: 11'].replace(r'[\$,]', '', regex=True), errors='coerce').fillna(0) +
    pd.to_numeric(DF_Nomina_Agricultura_NETA['Unnamed: 12'].replace(r'[\$,]', '', regex=True), errors='coerce').fillna(0) +
    pd.to_numeric(DF_Nomina_Agricultura_NETA['Unnamed: 13'].replace(r'[\$,]', '', regex=True), errors='coerce').fillna(0) +
    pd.to_numeric(DF_Nomina_Agricultura_NETA['Unnamed: 15'].replace(r'[\$,]', '', regex=True), errors='coerce').fillna(0)
)


# 4. LIMPIAR Y CONVERTIR TODAS LAS COLUMNAS NUMÉRICAS
# Quitar símbolos $ y comas, convertir a números
columnas_numericas = [
    'Salario_Bruto',
    'Descuento_AFP',
    'Descuento_ISR',
    'Descuento_SFS',
    'Total_Descuentos',
    'Salario_Neto'
]


for col in columnas_numericas:
    DF_Nomina_Agricultura_NETA[col] = pd.to_numeric(
        DF_Nomina_Agricultura_NETA[col].astype(str).replace(r'[\$,]', '', regex=True),
        errors='coerce'
    )


# 5. Seleccionar SOLO las 11 columnas finales
DF_Nomina_Agricultura_NETA = DF_Nomina_Agricultura_NETA[[
    'Nombre_Completo',
    'Cargo_Funcion',
    'Locacion_Departamento',
    'Salario_Bruto',
    'Descuento_AFP',
    'Descuento_ISR',
    'Descuento_SFS',
    'Otros_Descuentos',
    'Total_Descuentos',  # ← Ahora será un número limpio
    'Salario_Neto'
]]
```

```python
DF_Nomina_Agricultura_NETA
```

```python
DF_Nomina_Agricultura_NETA.head(51)
```

```python
# 1. Eliminar específicamente la fila que tiene "Nombres Apellidos" en Nombre_Completo
DF_Nomina_Agricultura_NETA = DF_Nomina_Agricultura_NETA[DF_Nomina_Agricultura_NETA['Nombre_Completo'] != 'Nombres Apellidos']

# 2. Eliminar líneas con cualquier valor NaN
DF_Nomina_Agricultura_NETA = DF_Nomina_Agricultura_NETA.dropna(how='any')

# 3. Eliminar duplicados de 'Nombre_Completo' manteniendo el mayor Salario_Neto
DF_Nomina_Agricultura_NETA = DF_Nomina_Agricultura_NETA.sort_values('Salario_Neto', ascending=False)
DF_Nomina_Agricultura_NETA = DF_Nomina_Agricultura_NETA.drop_duplicates(subset=['Nombre_Completo'], keep='first')

# 4. Resetear índice
DF_Nomina_Agricultura_NETA = DF_Nomina_Agricultura_NETA.reset_index(drop=True)
```

```python
DF_Nomina_ObrasPublicas_Bruta = pd.read_csv("ObrasPublicas_Nomina.csv")
```

```python
DF_Nomina_ObrasPublicas_Bruta
```

```python
DF_Nomina_ObrasPublicas_Bruta.head(51)
```

```python
# 1. Crear copia de la tabla original
DF_Nomina_ObrasPublicas_NETA = DF_Nomina_ObrasPublicas_Bruta.copy()

# 2. Renombrar columnas (por posición, no por nombre específico)
column_renames = {
    DF_Nomina_ObrasPublicas_NETA.columns[0]: 'Nombre_Completo',
    DF_Nomina_ObrasPublicas_NETA.columns[1]: 'Cargo_Funcion',
    DF_Nomina_ObrasPublicas_NETA.columns[2]: 'Locacion_Departamento',
    DF_Nomina_ObrasPublicas_NETA.columns[4]: 'Salario_Bruto',
    DF_Nomina_ObrasPublicas_NETA.columns[5]: 'Descuento_AFP',
    DF_Nomina_ObrasPublicas_NETA.columns[6]: 'Descuento_ISR',
    DF_Nomina_ObrasPublicas_NETA.columns[7]: 'Descuento_SFS',
    DF_Nomina_ObrasPublicas_NETA.columns[8]: 'Otros_Descuentos',
    DF_Nomina_ObrasPublicas_NETA.columns[9]: 'Salario_Neto'
}

DF_Nomina_ObrasPublicas_NETA = DF_Nomina_ObrasPublicas_NETA.rename(columns=column_renames)

# 3. Función para limpiar números
def limpiar_numero(valor):
    if isinstance(valor, str):
        valor_limpio = valor.replace('$', '').replace(',', '').strip()
        try:
            return float(valor_limpio)
        except:
            return 0.0
    elif pd.isna(valor):
        return 0.0
    else:
        return float(valor)

# 4. Limpiar columnas numéricas
columnas_numericas = ['Salario_Bruto', 'Descuento_AFP', 'Descuento_ISR',
                      'Descuento_SFS', 'Otros_Descuentos', 'Salario_Neto']

for col in columnas_numericas:
    DF_Nomina_ObrasPublicas_NETA[col] = DF_Nomina_ObrasPublicas_NETA[col].apply(limpiar_numero)

# 5. Calcular Total_Descuentos
DF_Nomina_ObrasPublicas_NETA['Total_Descuentos'] = (
    DF_Nomina_ObrasPublicas_NETA['Descuento_AFP'] +
    DF_Nomina_ObrasPublicas_NETA['Descuento_ISR'] +
    DF_Nomina_ObrasPublicas_NETA['Descuento_SFS'] +
    DF_Nomina_ObrasPublicas_NETA['Otros_Descuentos']
)

# 6. Seleccionar columnas finales
columnas_finales = [
    'Nombre_Completo',
    'Cargo_Funcion',
    'Locacion_Departamento',
    'Salario_Bruto',
    'Descuento_AFP',
    'Descuento_ISR',
    'Descuento_SFS',
    'Otros_Descuentos',
    'Total_Descuentos',
    'Salario_Neto'
]

DF_Nomina_ObrasPublicas_NETA = DF_Nomina_ObrasPublicas_NETA[columnas_finales]

# 7. Asegurar tipos de datos
DF_Nomina_ObrasPublicas_NETA['Nombre_Completo'] = DF_Nomina_ObrasPublicas_NETA['Nombre_Completo'].astype(str)
DF_Nomina_ObrasPublicas_NETA['Cargo_Funcion'] = DF_Nomina_ObrasPublicas_NETA['Cargo_Funcion'].astype(str)
DF_Nomina_ObrasPublicas_NETA['Locacion_Departamento'] = DF_Nomina_ObrasPublicas_NETA['Locacion_Departamento'].astype(str)

for col in ['Salario_Bruto', 'Descuento_AFP', 'Descuento_ISR', 'Descuento_SFS',
            'Otros_Descuentos', 'Total_Descuentos', 'Salario_Neto']:
    DF_Nomina_ObrasPublicas_NETA[col] = DF_Nomina_ObrasPublicas_NETA[col].astype('float64')

```

```python
DF_Nomina_ObrasPublicas_NETA
```

```python
# VERSIÓN CON VERIFICACIÓN

filas_iniciales = len(DF_Nomina_ObrasPublicas_NETA)

# 1. Eliminar filas no deseadas
mascara = (
    DF_Nomina_ObrasPublicas_NETA['Nombre_Completo'].astype(str).str.contains('DEPARTAMENTO DE NOMINA', case=False, na=False) |
    DF_Nomina_ObrasPublicas_NETA['Nombre_Completo'].astype(str).str.contains('RELACION DE EMPLEADOS', case=False, na=False) |
    DF_Nomina_ObrasPublicas_NETA['Nombre_Completo'].astype(str).str.contains('Empleado (Nombres y Apellidos)', case=False, na=False)
)
DF_Nomina_ObrasPublicas_NETA = DF_Nomina_ObrasPublicas_NETA[~mascara]
eliminados_paso1 = filas_iniciales - len(DF_Nomina_ObrasPublicas_NETA)

# 2. Eliminar nulos
antes_nulos = len(DF_Nomina_ObrasPublicas_NETA)
DF_Nomina_ObrasPublicas_NETA = DF_Nomina_ObrasPublicas_NETA.dropna()
eliminados_paso2 = antes_nulos - len(DF_Nomina_ObrasPublicas_NETA)

# 3. Eliminar duplicados
antes_dups = len(DF_Nomina_ObrasPublicas_NETA)
DF_Nomina_ObrasPublicas_NETA = DF_Nomina_ObrasPublicas_NETA.sort_values('Salario_Neto', ascending=False)
DF_Nomina_ObrasPublicas_NETA = DF_Nomina_ObrasPublicas_NETA.drop_duplicates(subset=['Nombre_Completo'], keep='first')
DF_Nomina_ObrasPublicas_NETA = DF_Nomina_ObrasPublicas_NETA.reset_index(drop=True)
eliminados_paso3 = antes_dups - len(DF_Nomina_ObrasPublicas_NETA)

print(f"Resumen de limpieza:")
print(f"- Filas iniciales: {filas_iniciales}")
print(f"- Eliminadas (encabezados): {eliminados_paso1}")
print(f"- Eliminadas (nulos): {eliminados_paso2}")
print(f"- Eliminadas (duplicados): {eliminados_paso3}")
print(f"- Filas finales: {len(DF_Nomina_ObrasPublicas_NETA)}")

```

```python
# 1. Crear una copia del DataFrame original para no modificarlo
DF_Nomina_Aduanas_NETA = DF_Nomina_Aduanas_Bruta.copy()

# 2. Eliminar las primeras filas que contienen información de encabezado
DF_Nomina_Aduanas_NETA = DF_Nomina_Aduanas_NETA.iloc[8:].reset_index(drop=True)

# 3. Renombrar las columnas según el esquema original
DF_Nomina_Aduanas_NETA.columns = [
    'Cantidad', 'Nombre_Completo', 'Cargo_Funcion', 'Locacion_Departamento',
    'Genero', 'Tipo_Empleado', 'Salario_Bruto', 'Descuento_AFP', 'Descuento_SFS',
    'Descuento_ISR', 'Otros_Descuentos', 'Salario_Neto'
]

# 4. Seleccionar solo las columnas que necesitas
columnas_necesarias = [
    'Nombre_Completo', 'Cargo_Funcion', 'Locacion_Departamento',
    'Salario_Bruto', 'Descuento_AFP', 'Descuento_ISR', 'Descuento_SFS',
    'Otros_Descuentos', 'Salario_Neto'
]

DF_Nomina_Aduanas_NETA = DF_Nomina_Aduanas_NETA[columnas_necesarias]

# 5. Calcular la columna Total_Descuentos sumando todos los descuentos
DF_Nomina_Aduanas_NETA['Total_Descuentos'] = (
    DF_Nomina_Aduanas_NETA['Descuento_AFP'].astype(str).str.replace(',', '').astype(float) +
    DF_Nomina_Aduanas_NETA['Descuento_ISR'].astype(str).str.replace(',', '').astype(float) +
    DF_Nomina_Aduanas_NETA['Descuento_SFS'].astype(str).str.replace(',', '').astype(float) +
    DF_Nomina_Aduanas_NETA['Otros_Descuentos'].astype(str).str.replace(',', '').astype(float)
)

# 6. Convertir las columnas numéricas a float (eliminando comas de miles)
columnas_numericas = [
    'Salario_Bruto', 'Descuento_AFP', 'Descuento_ISR',
    'Descuento_SFS', 'Otros_Descuentos', 'Salario_Neto'
]

for col in columnas_numericas:
    DF_Nomina_Aduanas_NETA[col] = DF_Nomina_Aduanas_NETA[col].astype(str).str.replace(',', '').astype(float)

# 7. Reordenar las columnas en el orden solicitado
columnas_ordenadas = [
    'Nombre_Completo', 'Cargo_Funcion', 'Locacion_Departamento',
    'Salario_Bruto', 'Descuento_AFP', 'Descuento_ISR', 'Descuento_SFS',
    'Otros_Descuentos', 'Total_Descuentos', 'Salario_Neto'
]

DF_Nomina_Aduanas_NETA = DF_Nomina_Aduanas_NETA[columnas_ordenadas]

# 8. Convertir las columnas de texto a string (object) y limpiar espacios
columnas_texto = ['Nombre_Completo', 'Cargo_Funcion', 'Locacion_Departamento']
for col in columnas_texto:
    DF_Nomina_Aduanas_NETA[col] = DF_Nomina_Aduanas_NETA[col].astype(str).str.strip()

```

```python
# 1. Eliminar todas las líneas que contengan al menos 1 o más nulos, nan o NaN
DF_Nomina_Aduanas_NETA = DF_Nomina_Aduanas_NETA.dropna(how='any')

# 2. Eliminar strings vacíos (primero identificamos y luego eliminamos)
# Para cada columna de tipo object, convertimos strings vacíos a NaN
for col in DF_Nomina_Aduanas_NETA.select_dtypes(include='object').columns:
    DF_Nomina_Aduanas_NETA[col] = DF_Nomina_Aduanas_NETA[col].replace('', np.nan)
    # También eliminar strings que solo contienen espacios
    DF_Nomina_Aduanas_NETA[col] = DF_Nomina_Aduanas_NETA[col].replace(r'^\s*$', np.nan, regex=True)

# Eliminar filas que ahora tienen NaN después de convertir strings vacíos
DF_Nomina_Aduanas_NETA = DF_Nomina_Aduanas_NETA.dropna(how='any')

# 3. Eliminar duplicados por "Nombre_Completo" manteniendo el mayor "Salario_Neto"
DF_Nomina_Aduanas_NETA = DF_Nomina_Aduanas_NETA.sort_values('Salario_Neto', ascending=False)
DF_Nomina_Aduanas_NETA = DF_Nomina_Aduanas_NETA.drop_duplicates(subset='Nombre_Completo', keep='first')

# Opcional: Reordenar el índice si lo deseas
DF_Nomina_Aduanas_NETA = DF_Nomina_Aduanas_NETA.reset_index(drop=True)
```

```python
# Crear una copia del DataFrame original para trabajar
DF_Nomina_MedioAmbiente_NETA = DF_Nomina_MedioAmbiente_Bruto.copy()

# 1. Filtrar solo las filas que tienen datos de empleados (filas con números en la primera columna)
# Convertir la primera columna a numérico donde sea posible
DF_Nomina_MedioAmbiente_NETA['Unnamed: 0'] = pd.to_numeric(DF_Nomina_MedioAmbiente_NETA['Unnamed: 0'], errors='coerce')

# Filtrar solo las filas donde la primera columna es numérica (son los registros de empleados)
DF_Nomina_MedioAmbiente_NETA = DF_Nomina_MedioAmbiente_NETA[DF_Nomina_MedioAmbiente_NETA['Unnamed: 0'].notna()]

# 2. Seleccionar y renombrar las columnas necesarias
DF_Nomina_MedioAmbiente_NETA = DF_Nomina_MedioAmbiente_NETA.rename(columns={
    'Unnamed: 1': 'Nombre_Completo',
    'Unnamed: 2': 'Cargo_Funcion',
    'Unnamed: 5': 'Salario_Bruto',
    'Unnamed: 6': 'Descuento_AFP',
    'Unnamed: 7': 'Descuento_ISR',
    'Unnamed: 8': 'Descuento_SFS',
    'Unnamed: 9': 'Otros_Descuentos',
    'Unnamed: 10': 'Total_Descuentos',
    'Unnamed: 11': 'Salario_Neto'
})

# 3. Crear la columna Locacion_Departamento con el valor constante
DF_Nomina_MedioAmbiente_NETA['Locacion_Departamento'] = 'Ministerio de Medio Ambiente'

# 4. Seleccionar solo las columnas que necesitamos en el orden correcto
columnas_necesarias = [
    'Nombre_Completo',
    'Cargo_Funcion',
    'Locacion_Departamento',
    'Salario_Bruto',
    'Descuento_AFP',
    'Descuento_ISR',
    'Descuento_SFS',
    'Otros_Descuentos',
    'Total_Descuentos',
    'Salario_Neto'
]

DF_Nomina_MedioAmbiente_NETA = DF_Nomina_MedioAmbiente_NETA[columnas_necesarias]

# 5. Convertir las columnas numéricas a float64
# Primero, limpiar los valores numéricos (eliminar comas y espacios)
columnas_numericas = ['Salario_Bruto', 'Descuento_AFP', 'Descuento_ISR',
                      'Descuento_SFS', 'Otros_Descuentos', 'Total_Descuentos', 'Salario_Neto']

for col in columnas_numericas:
    # Convertir a string, eliminar comas y espacios, luego a float
    DF_Nomina_MedioAmbiente_NETA[col] = DF_Nomina_MedioAmbiente_NETA[col].astype(str).str.replace(',', '').str.strip()
    DF_Nomina_MedioAmbiente_NETA[col] = pd.to_numeric(DF_Nomina_MedioAmbiente_NETA[col], errors='coerce')

# 6. Convertir las columnas de texto a object (string)
columnas_texto = ['Nombre_Completo', 'Cargo_Funcion', 'Locacion_Departamento']
for col in columnas_texto:
    DF_Nomina_MedioAmbiente_NETA[col] = DF_Nomina_MedioAmbiente_NETA[col].astype('object')
```

```python
# 1. Crear una copia del DataFrame original para preservarlo intacto
DF_Nomina_ImpuestosInternos_Neta = DF_Nomina_ImpuestosInternos_Bruto.copy()

# 2. Renombrar las columnas según tus especificaciones
DF_Nomina_ImpuestosInternos_Neta = DF_Nomina_ImpuestosInternos_Neta.rename(columns={
    'Unnamed: 0': 'Nombre_Completo',
    'Unnamed: 2': 'Cargo_Funcion',
    'Unnamed: 3': 'Locacion_Departamento',
    'Unnamed: 6': 'Salario_Bruto',
    'Unnamed: 8': 'Descuento_AFP',
    'Unnamed: 7': 'Descuento_ISR',
    'Unnamed: 9': 'Descuento_SFS',
    'Unnamed: 10': 'Otros_Descuentos',
    'Unnamed: 11': 'Salario_Neto'
})

# 3. Seleccionar solo las columnas que necesitamos (eliminar las que no)
columnas_necesarias = [
    'Nombre_Completo', 'Cargo_Funcion', 'Locacion_Departamento',
    'Salario_Bruto', 'Descuento_AFP', 'Descuento_ISR',
    'Descuento_SFS', 'Otros_Descuentos', 'Salario_Neto'
]

DF_Nomina_ImpuestosInternos_Neta = DF_Nomina_ImpuestosInternos_Neta[columnas_necesarias]

# 4. IDENTIFICAR Y ELIMINAR FILAS DE ENCABEZADO/NO DATOS
# Las filas que contienen texto como "Salario Bruto\n(RD$)" son encabezados
# Primero, eliminamos filas donde 'Nombre_Completo' es NaN o está vacío
DF_Nomina_ImpuestosInternos_Neta = DF_Nomina_ImpuestosInternos_Neta.dropna(subset=['Nombre_Completo'])

# Segundo, eliminamos filas que contienen texto de encabezado en la columna 'Nombre_Completo'
textos_encabezado = [
    'Nombre (s) y Apellido (s)',
    'Salario Bruto\n(RD$)',
    'ISR\n(Ley 11-92',
    'Seguridad Social (Ley 87-01)',
    'Otros\nDescuentos',
    'Salario Neto\n(RD$)',
    'Seguro de Pension',
    'Seguro de Salud',
    'Empleado\n(2.87%)',
    'Empleado\n(3.04%)',
    ''
]

# Filtrar para mantener solo filas que NO contienen textos de encabezado
mask = ~DF_Nomina_ImpuestosInternos_Neta['Nombre_Completo'].isin(textos_encabezado)
DF_Nomina_ImpuestosInternos_Neta = DF_Nomina_ImpuestosInternos_Neta[mask]

# 5. Resetear el índice después de eliminar filas
DF_Nomina_ImpuestosInternos_Neta = DF_Nomina_ImpuestosInternos_Neta.reset_index(drop=True)

# 6. Limpiar y convertir las columnas numéricas
columnas_numericas = ['Salario_Bruto', 'Descuento_AFP', 'Descuento_ISR',
                      'Descuento_SFS', 'Otros_Descuentos', 'Salario_Neto']

for col in columnas_numericas:
    # Convertir a string, eliminar comas, reemplazar valores vacíos por NaN
    DF_Nomina_ImpuestosInternos_Neta[col] = (
        DF_Nomina_ImpuestosInternos_Neta[col]
        .astype(str)
        .str.replace(',', '', regex=False)
        .replace('nan', np.nan)  # Reemplazar 'nan' string por NaN
        .replace('', np.nan)     # Reemplazar strings vacíos por NaN
    )

    # Convertir a float, forzando errores a NaN
    DF_Nomina_ImpuestosInternos_Neta[col] = pd.to_numeric(
        DF_Nomina_ImpuestosInternos_Neta[col],
        errors='coerce'
    )

# 7. Verificar que tenemos datos numéricos válidos
print("Verificación de conversión numérica:")
for col in columnas_numericas:
    print(f"{col}: {DF_Nomina_ImpuestosInternos_Neta[col].notna().sum()} valores no nulos")

# 8. Calcular la columna Total_Descuentos
DF_Nomina_ImpuestosInternos_Neta['Total_Descuentos'] = (
    DF_Nomina_ImpuestosInternos_Neta['Descuento_AFP'].fillna(0) +
    DF_Nomina_ImpuestosInternos_Neta['Descuento_ISR'].fillna(0) +
    DF_Nomina_ImpuestosInternos_Neta['Descuento_SFS'].fillna(0) +
    DF_Nomina_ImpuestosInternos_Neta['Otros_Descuentos'].fillna(0)
)

# 9. Reordenar las columnas en el orden especificado
column_order = [
    'Nombre_Completo', 'Cargo_Funcion', 'Locacion_Departamento',
    'Salario_Bruto', 'Descuento_AFP', 'Descuento_ISR',
    'Descuento_SFS', 'Otros_Descuentos', 'Total_Descuentos', 'Salario_Neto'
]

DF_Nomina_ImpuestosInternos_Neta = DF_Nomina_ImpuestosInternos_Neta[column_order]

# 10. Convertir los tipos de datos finales
# Columnas de tipo object (categóricas)
DF_Nomina_ImpuestosInternos_Neta['Nombre_Completo'] = DF_Nomina_ImpuestosInternos_Neta['Nombre_Completo'].astype('object')
DF_Nomina_ImpuestosInternos_Neta['Cargo_Funcion'] = DF_Nomina_ImpuestosInternos_Neta['Cargo_Funcion'].astype('object')
DF_Nomina_ImpuestosInternos_Neta['Locacion_Departamento'] = DF_Nomina_ImpuestosInternos_Neta['Locacion_Departamento'].astype('object')

# Columnas de tipo float64 (numéricas) - ya están convertidas
# Solo asegurarnos del tipo
for col in ['Salario_Bruto', 'Descuento_AFP', 'Descuento_ISR',
            'Descuento_SFS', 'Otros_Descuentos', 'Total_Descuentos', 'Salario_Neto']:
    if col in DF_Nomina_ImpuestosInternos_Neta.columns:
        DF_Nomina_ImpuestosInternos_Neta[col] = DF_Nomina_ImpuestosInternos_Neta[col].astype('float64')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score

# 1. Crear y entrenar el modelo OLS (LinearRegression usa OLS por defecto)
model_ols = LinearRegression()
model_ols.fit(X_train, y_train)

# 2. Hacer predicciones
y_predict = model_ols.predict(X_test)

# 3. Crear DataFrame con resultados
df_results = pd.concat([
    y_test.reset_index(drop=True),
    pd.Series(y_predict, name="Predicted_Salario_Bruto")
], axis=1)

# 4. Calcular métricas
mae = mean_absolute_error(y_test, y_predict)
rmse = root_mean_squared_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)


print(f"Modelo entrenado con {len(X_train)} muestras")
print(f"Evaluado con {len(X_test)} muestras")

print("\n MÉTRICAS DE EVALUACIÓN:")
print(f"• Mean Absolute Error (MAE): {mae:,.2f}")
print(f"• Root Mean Squared Error (RMSE): {rmse:,.2f}")
print(f"• R² Score: {r2:.4f}")

print("\n")
print(f"• MAE: Error promedio de {mae:,.2f} pesos por predicción")
print(f"• RMSE: {rmse:,.2f} pesos (penaliza errores grandes)")
print(f"• R²: {r2:.2%} de la varianza en Salario_Bruto es explicada por el modelo")
```
